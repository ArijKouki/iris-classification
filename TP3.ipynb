Par conséquent, lorsque le modèle effectue des prédictions sur ces 50 observations, il n'a pas été entraîné sur des exemples de cette classe particulière et ne peut pas être en mesure de faire des prédictions précises pour elle.

Pour équilibrer, on pourrait utiliser une méthode de fractionnement plus équilibrée, comme train_test_split de scikit-learn, pour diviser l'ensemble de données en ensembles d'apprentissage et de test de manière aléatoire tout en préservant la proportion de classes.

### 2. Performances en généralisation


#### 2.1. Estimer l'erreur réelle par division de l'échantillon d'apprentissage


```python
import numpy as np

def split(s):
    data = s.data
    targets = s.target

    combined = np.column_stack((data, targets))
    np.random.shuffle(combined)
    data = combined[:, :-1]
    targets = combined[:, -1]

    n = int(len(data) * 2 / 3)
    dataS1 = data[:n]
    targetS1 = targets[:n]
    dataS2 = data[n:]
    targetS2 = targets[n:]

    return dataS1, targetS1, dataS2, targetS2

dataS1, targetS1, dataS2, targetS2 = split(irisData)
print("dataS1: ", dataS1)
print("targetS1: ", targetS1)
print("dataS2: ", dataS2)
print("targetS2: ", targetS2)
```

```python
def test(S, clf):
    dataS1, targetS1, dataS2, targetS2 = split(S)

    clf.fit(dataS1, targetS1)

    predictions_train = clf.predict(dataS1)
    error_rate_train = np.sum(predictions_train != targetS1) / len(targetS1)

    predictions_test = clf.predict(dataS2)
    error_rate_test = np.sum(predictions_test != targetS2) / len(targetS2)

    return error_rate_train, error_rate_test
```

```python
nb = naive_bayes.MultinomialNB(fit_prior=True)

apparent_error = ea_method_1
estimated_error = test(irisData, nb)[1]

print("Apparent Error:", apparent_error)
print("Estimated Error:", estimated_error)
```

On n'obtient généralement pas toujours la même estimation pour l'erreur réelle, car elle dépend de la manière dont les données sont réparties entre les ensembles d'apprentissage et de test.

Pour obtenir une estimation plus robuste de la performance du modèle, on pourrait répéter le processus de division et d'estimation plusieurs fois, puis calculer la moyenne ou d'autres statistiques récapitulatives de l'erreur estimée. Cela permettrait d'obtenir une perspective plus stable de la capacité de généralisation du modèle.

```python
def average_error(cls,t, num_repeats):


    average_train_errors = []
    average_test_errors = []

    for _ in range(num_repeats):
        train_errors = []
        test_errors = []
        for _ in range(t):
            error_train,error_test = test(irisData, cls)
            train_errors.append(error_train)
            test_errors.append(error_test)

        average_train_error = np.mean(train_errors)
        average_train_errors.append(average_train_error)

        average_test_error = np.mean(test_errors)
        average_test_errors.append(average_test_error)

    return np.mean(average_train_errors), np.mean(average_test_errors)

nb = naive_bayes.MultinomialNB(fit_prior=True)

t_values = [10, 50, 100, 200, 500, 1000]

results_df3 = pd.DataFrame(columns=['t', 'average_train_error', 'average_test_error'])

for t in t_values:
    avg_train_error, avg_test_error = average_error(nb,t, num_repeats=20)
    results_df3 = pd.concat([results_df3, pd.DataFrame({'t': [t], 'average_train_error': [avg_train_error],'average_test_error':[avg_test_error]})], ignore_index=True)

results_df3
```

Globalement, on observe une certaine stabilité dans l'erreur moyenne, avec des variations modestes. Cependant, il est important de noter que la stabilité peut être relative. Ces variations peuvent être dues à la nature stochastique de la division des données et à la taille relativement petite du jeu de données.

```python
import numpy as np

def split10(s):
    data = s.data
    targets = s.target

    combined = np.column_stack((data, targets))
    np.random.shuffle(combined)
    data = combined[:, :-1]
    targets = combined[:, -1]

    n = int(len(data) * 9 /   10)
    dataS1 = data[:n]
    targetS1 = targets[:n]
    dataS2 = data[n:]
    targetS2 = targets[n:]

    return dataS1, targetS1, dataS2, targetS2

def test10(S, clf):
    dataS1, targetS1, dataS2, targetS2 = split10(S)
    clf.fit(dataS1, targetS1)

    predictions_train = clf.predict(dataS1)
    error_rate_train = np.sum(predictions_train != targetS1) / len(targetS1)

    predictions_test = clf.predict(dataS2)
    error_rate_test = np.sum(predictions_test != targetS2) / len(targetS2)

    return error_rate_train, error_rate_test

def average_error10(cls,t, num_repeats):

    average_train_errors = []
    average_test_errors = []

    for _ in range(num_repeats):
        train_errors = []
        test_errors = []
        for _ in range(t):
            error_train,error_test = test10(irisData, cls)
            train_errors.append(error_train)
            test_errors.append(error_test)

        average_train_error = np.mean(train_errors)
        average_train_errors.append(average_train_error)

        average_test_error = np.mean(test_errors)
        average_test_errors.append(average_test_error)

    return np.mean(average_train_errors), np.mean(average_test_errors)

nb = naive_bayes.MultinomialNB(fit_prior=True)
t_values = [10, 50, 100, 200, 500, 1000]
results_df10 = pd.DataFrame(columns=['t', 'average_train_error', 'average_test_error'])

for t in t_values:
    avg_train_error, avg_test_error = average_error10(nb,t, num_repeats=20)
    results_df10 = pd.concat([results_df10, pd.DataFrame({'t': [t], 'average_train_error': [avg_train_error],'average_test_error':[avg_test_error]})], ignore_index=True)

results_df10
```

En comparant les résultats, on peut observer que l'erreur estimée avec un échantillon de test représentant 1/10 de l'échantillon initial est un peu plus faible que celle avec une proportion d'1/3.

En général, une plus grande taille d'échantillon de test (1/3) est préférable pour obtenir une estimation plus stable et précise de l'erreur de généralisation.

```python
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
import pandas as pd
import numpy as np

def evaluate(cls):
  test_sizes = [0.2, 0.25, 0.3, 0.33, 0.4]
  results_spl = pd.DataFrame(columns=['Test Size', 'Train Error Rate','Test Error Rate'])

  for test_size in test_sizes:
      D_train, D_test, C_train, C_test = train_test_split(irisData.data, irisData.target, test_size=test_size, random_state=42)

      cls.fit(D_train, C_train)

      predictions_train = cls.predict(D_train)
      error_rate_train = np.sum(predictions_train != C_train) / len(C_train)

      predictions_test = cls.predict(D_test)
      error_rate_test = np.sum(predictions_test != C_test) / len(C_test)

      results_spl = pd.concat([results_spl, pd.DataFrame({'Test Size': [test_size], 'Train Error Rate': [error_rate_train],'Test Error Rate': [error_rate_test]})], ignore_index=True)
  return results_spl

nb = naive_bayes.MultinomialNB(fit_prior=True)
results_spl=evaluate(nb)
results_spl
```

#### 2.2 Estimer l'erreur réelle par validation croisée


```python
import numpy as np

def k_fold_cross_validation(X, y, classifier, k=5):
    """
    Perform k-fold cross-validation for a given classifier.

    Parameters:
    - X: Features (numpy array or pandas DataFrame)
    - y: Labels (numpy array or pandas Series)
    - classifier: Classifier object with fit and predict methods
    - k: Number of folds for cross-validation

    Returns:
    - List of error rates for each fold
    """
    n = len(X)
    fold_size = n // k
    error_rates = []

    for i in range(k):
        start, end = i * fold_size, (i + 1) * fold_size

        X_train = np.concatenate((X[:start], X[end:]), axis=0)
        y_train = np.concatenate((y[:start], y[end:]), axis=0)
        X_test = X[start:end]
        y_test = y[start:end]

        classifier.fit(X_train, y_train)

        predictions = classifier.predict(X_test)

        error_rate = np.sum(predictions != y_test) / len(y_test)
        error_rates.append(error_rate)

    return error_rates
```

```python
nb = naive_bayes.MultinomialNB(fit_prior=True)

error_rates = k_fold_cross_validation(irisData.data, irisData.target, nb, k=5)

for i, error_rate in enumerate(error_rates):
    print(f"Fold {i+1} Error Rate: {error_rate}")

average_error_rate = np.mean(error_rates)
print(f"Average Error Rate: {average_error_rate}")
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import naive_bayes

nb = naive_bayes.MultinomialNB(fit_prior=True)

k_values = [2, 3, 5, 8, 10]
results_cv = pd.DataFrame(columns=['k', 'Error Rate'])


for k in k_values:
    scores = cross_val_score(nb, irisData.data, irisData.target, cv=k)

    error_rates = 1 - scores

    for i, error_rate in enumerate(error_rates):
        print(f"k={k}, Fold {i+1} Error Rate: {error_rate}")

    average_error_rate = np.mean(error_rates)
    print(f"k={k}, Average Error Rate: {average_error_rate}")
    results_cv = pd.concat([results_cv, pd.DataFrame({'k': [k], 'Error Rate': [average_error_rate]})], ignore_index=True)
    print("-" * 30)

results_cv
```

```python
# using cross_validate to get train
from sklearn.model_selection import cross_validate
from sklearn import naive_bayes

def evaluate_cv(cls):
  k_values = [2, 3, 5, 8, 10]
  results_cv = pd.DataFrame(columns=['k', 'Train Error', 'Test Error'])

  for k in k_values:
      cv_results = cross_validate(cls, irisData.data, irisData.target, cv=k, return_train_score=True)

      train_errors = 1 - cv_results['train_score']
      test_errors = 1 - cv_results['test_score']

      """for i, (train_error, test_error) in enumerate(zip(train_errors, test_errors)):
          print(f"k={k}, Fold {i+1} Train Error: {train_error}, Test Error: {test_error}")"""

      average_train_error = np.mean(train_errors)
      average_test_error = np.mean(test_errors)

      #print(f"k={k}, Average Train Error: {average_train_error}, Average Test Error: {average_test_error}")
      results_cv = pd.concat([results_cv, pd.DataFrame({'k': [k], 'Train Error': [average_train_error], 'Test Error': [average_test_error]})], ignore_index=True)
      #print("-" * 30)
  return results_cv

nb = naive_bayes.MultinomialNB(fit_prior=True)
results_cv = evaluate_cv(nb)
results_cv
```
